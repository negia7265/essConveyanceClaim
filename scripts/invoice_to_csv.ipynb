### To train the modeL , dataframes of all the invoice are required.This notebook fetch all the invoice data from it's images and save it for further training.

```python
import cv2
import pytesseract
import time
import os
import numpy as np
import pandas 
```

```python
def preprocess_img(img):
    # Decode and convert to grayscale
    img = cv2.imdecode(np.frombuffer(img, np.uint8), cv2.IMREAD_GRAYSCALE)
    # Denoising Image
    img = cv2.fastNlMeansDenoising( img, None, 15, 7, 21 )   
    # Image Binarization
    img=cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    # Perform morphological operations (erosion and dilation)
    kernel = np.ones((1, 1), np.uint8)
    return cv2.morphologyEx(img, cv2.MORPH_DILATE, kernel)
```

```python
def img2csv_data(dir_name):
    start_time=time.time()
    csv_dir=dir_name+"_CSV"     
    image_dir=dir_name+"_IMAGES"
    os.mkdir(csv_dir)
    invoice_dirs=os.listdir(image_dir) # list all the invoice directories   
    for invoice_dir in invoice_dirs:  
        os.mkdir(f'{csv_dir}/{invoice_dir}')
        invoice_images=os.listdir(f"{image_dir}/{invoice_dir}")  
        for invoice in invoice_images:      
            num=invoice.split('.')[0]
            f=open(f"{image_dir}/{invoice_dir}/{invoice}",'rb')
            df=pytesseract.image_to_data(preprocess_img(f.read()),lang='eng_ruppe',output_type='data.frame')
            df.to_csv(f'{csv_dir}/{invoice_dir}/{num}.csv')
    end_time=time.time()
    print(f'converted invoice images to csv data in {round(end_time-start_time,2)} seconds')
```
